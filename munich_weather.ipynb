```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv("weather_data_munich.csv", index_col="date")
```

```python
weather
```

```python
weather.loc["1/1/1979"] #single date
weather.loc["1/1/1979":"1/30/1979",:] # slice through a period of time
```

```python
new_weather = weather.loc["5/19/1992":"11/9/2022",:]
```

```python
new_weather
```

```python
new_weather.apply(pd.isnull).sum()/new_weather.shape[0]*100
```

snow we will replace with 0 since most of the NaN values are located in spring, summer or autum season
wdir
wspd
wpgt
tsun will replace NaN values with 'fill' method.

```python
clean_weather = new_weather.fillna(0)
```

```python
clean_weather
```

```python
clean_weather.apply(pd.isnull).sum()/clean_weather.shape[0]
```

creating a datafram snow_data we replace all values with the fill method. We then replace the snow colum from the snow_data dataframe into the clean_weather snow colum

```python
snow_data = new_weather.fillna(method = "ffill")
```

```python
clean_weather["snow"] = snow_data["snow"]
```

```python
clean_weather
```

```python
clean_weather.apply(pd.isnull).sum()/clean_weather.shape[0]*100
```

Now we have a perfectly clean data set and we are ready to train our model

```python
clean_weather.index = pd.to_datetime(clean_weather.index)
```

```python
clean_weather.plot(y="tmax")
```

```python
clean_weather["target"] = clean_weather.shift(-1)["snow"]
```

```python
clean_weather
```

```python
clean_weather.iloc[11119,10] = 0
clean_weather
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.1)
```

```python
predictors = ["tavg","tmax","tmin","prcp","snow","wdir","wspd","wpgt","pres","tsun"]
```

```python
train = clean_weather.loc[:"2020-12-31"]
```

```python
test = clean_weather.loc["2021-01-01":"2021-12-25"]
test
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"], predictions)
```

```python
reg.coef_
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis = 1)
```

```python
combined.columns= ["actual","predictions"]
combined
```

```python
combined.plot()
```
